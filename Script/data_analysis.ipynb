```python
%matplotlib notebook
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
import scipy.stats as st
```

```python
input_path = os.path.join("..","..","Source")
output_path = os.path.join("..","Graphs")
```

```python
movies = pd.read_csv(input_path +'/IMDb_movies.csv')
names = pd.read_csv(input_path +'/IMDb_names.csv')
ratings = pd.read_csv(input_path +'/IMDb_ratings.csv')
title_principals = pd.read_csv(input_path +'/IMDb_title_principals.csv')
display(movies.head())
display(names.head())
display(ratings.head())
display(title_principals.head())
```

```python
movies_columns=['imdb_title_id','title','original_title','year','date_published','genre','duration','country','language','director','writer','production_company','budget','usa_gross_income','worlwide_gross_income']
names_columns=['imdb_name_id','name','birth_name','height','birth_year','date_of_birth','place_of_birth','death_year','date_of_death','spouses','divorces','spouses_with_children','children']
title_principals_columns=['imdb_title_id','imdb_name_id','category']
ratings_columns=['imdb_title_id','weighted_average_vote','total_votes','mean_vote','median_vote','votes_10','votes_9','votes_8','votes_7','votes_6','votes_5','votes_4','votes_3','votes_2','votes_1','allgenders_0age_avg_vote','allgenders_0age_votes','allgenders_18age_avg_vote','allgenders_18age_votes','allgenders_30age_avg_vote','allgenders_30age_votes','allgenders_45age_avg_vote','allgenders_45age_votes','males_0age_avg_vote','males_0age_votes','males_18age_avg_vote','males_18age_votes','males_30age_avg_vote','males_30age_votes','males_45age_avg_vote','males_45age_votes','females_0age_avg_vote','females_0age_votes','females_18age_avg_vote','females_18age_votes','females_30age_avg_vote','females_30age_votes','females_45age_avg_vote','females_45age_votes']
```

```python
usefull_movies=movies[movies_columns].copy()
display(usefull_movies.head())
usefull_names=names[names_columns].copy()
display(usefull_names.head())
usefull_ratings=ratings[ratings_columns].copy()
display(usefull_ratings.head())
usefull_title_principals=title_principals[title_principals_columns].copy()
display(usefull_title_principals.head())
```

```python
merged_data = pd.DataFrame.merge(usefull_title_principals,usefull_movies,how='left',on='imdb_title_id')
merged_data = merged_data.merge(usefull_names,how='left',on='imdb_name_id')
merged_data = merged_data.merge(usefull_ratings,how='left',on='imdb_title_id')
display(merged_data)
```
